```python
import copy
import numpy as np
import cv2
from tensorflow.keras.models import load_model
```

```python
model1= r"C:\Users\user\Desktop\All Models\yuz.h5"
model = load_model(model1)
```

```python
path_f=r"C:\Users\user\Desktop\All Models\frontalface.xml"
face_cascade = cv2.CascadeClassifier(path_f) 
```

```python
face_cascade = cv2.CascadeClassifier(path_f)

path = r"C:\Users\user\Downloads\video_2024-01-24_19-16-11.mp4"
cap = cv2.VideoCapture(path)
emotion1 = []

while True:
    ret, frame = cap.read()
    if not ret:
        break  
    img = copy.deepcopy(frame)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        fc = img[y:y+h, x:x+w]
        roi = cv2.resize(fc, (256, 256))
        pred = model.predict(roi[np.newaxis, :, :])
        text_idx = np.argmax(pred)
        text_list = ['neutral', 'negative', 'positive']
        text = text_list[text_idx]
        emotion1.append(text)

nc1 = emotion1.count('neutral')
ngc1 = emotion1.count('negative')
pc1 = emotion1.count('positive')

print(f"neutral: {nc1}, negative: {ngc1}, positive: {pc1}")

cap.release()
cv2.destroyAllWindows()
```

```python
import moviepy.editor as mp
import speech_recognition as sr
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

output=[]

def extract_audio(path):
    video = mp.VideoFileClip(path)
    audio = video.audio
    audio.write_audiofile("audio.wav")

def transcribe_audio(audio_file):
    r = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio = r.record(source)
    try:
        text = r.recognize_google(audio)
        return text
    except sr.UnknownValueError:
        print("Audio could not be understood")
        return ""
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))
        return ""

# Function to analyze sentiment
def analyze_sentiment(text):
    sid = SentimentIntensityAnalyzer()  # Create a SentimentIntensityAnalyzer object
    blob = sid.polarity_scores(text)   # Use it to analyze sentiment
    sentiment = blob['compound']       # Access the sentiment score

    if sentiment > 0:
        emotion = "neutral"
    elif sentiment < 0: 
        emotion = "negative"
    else:
        emotion = "positive"

    return emotion

def split_text(text):
    words = text.split()  # Split text into words
    chunks = []
    current_chunk = []
    for word in words:
      if word:  # Check for non-empty words
        current_chunk.append(word)
        if len(current_chunk) == 25:
          chunks.append(" ".join(current_chunk))  # Join words with spaces
          current_chunk = []  # Reset the chunk
    if current_chunk:  # Add any remaining words
      chunks.append(" ".join(current_chunk))
    return chunks

output.append(text)
extract_audio(path)
text = transcribe_audio("audio.wav")

output1=[]

if text:
    text_chunks = split_text(text)  # Split the text into chunks

    for chunk in text_chunks:
        emotion = analyze_sentiment(chunk)  # Analyze sentiment for each chunk
        print("Chunk:", chunk)
        print("Detected emotion:", emotion)
        output1.append(emotion)
else:

    print("Transcription failed")



nc2 = output1.count('neutral')
ngc2 = output1.count('negative')
pc2 = output1.count('positive')

print(output1)

print(f"neutral: {nc2}, negative: {ngc2}, positive: {pc2}")
```
